```
# Inspired by https://towardsdatascience.com/understanding-political-twitter-ce3476a38377 to practice basic sentiment analysis, generating wordclouds
# Following this tutorial https://www.datacamp.com/community/tutorials/wordcloud-python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib
import seaborn as sns
! pip install wordcloud
! pip install Pillow
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
% matplotlib inline
from google.colab import files

```

```
# Import data file
clean = pd.read_csv('/content/clean.csv')
clean = pd.DataFrame(clean)
```

```
user = clean.groupby("username") # Group by username
ax = sns.countplot(x="username", data=clean) 
plt.savefig('count.png')
files.download('count.png')
```

We can see that AOC has the highest number of tweets, followed by Ben Shapiro, Ted Cruz, and then Pete Buttigieg.

```
# Test run only using one tweet from AOC
tweettxt = clean.tweet[0]
str(tweettxt)
wordcloud = WordCloud().generate(tweettxt)
# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.savefig('testwordcloud.png')
files.download('testwordcloud.png')
```

The test tweet works!

```
# ALL TWEETS IN MERGED
clean['tweet'] = clean['tweet'].astype(str)
text = " ".join(tweet for tweet in clean.tweet)
print ("There are {} words in the combination of all the tweets in the merged dataset.".format(len(text)))
wordcloud = WordCloud().generate(text)
# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.savefig('mergedwordcloud.png')
files.download('mergedwordcloud.png')
```

Now, we will iterate through the 4 people and generate wordclouds!

```
# Make combined strings for each user!
aoc = " ".join(tweet for tweet in clean[clean["username"]=="aoc"].tweet)
buttigieg = " ".join(tweet for tweet in clean[clean["username"]=="petebuttigieg"].tweet)
cruz = " ".join(tweet for tweet in clean[clean["username"]=="tedcruz"].tweet)
shapiro = " ".join(tweet for tweet in clean[clean["username"]=="benshapiro"].tweet)



wordcloud = WordCloud().generate(aoc)
# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.savefig('aoc.png')
files.download('aoc.png')

wordcloud = WordCloud().generate(buttigieg)
# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.savefig('buttigieg.png')
files.download('buttigieg.png')


wordcloud = WordCloud().generate(cruz)
# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.savefig('cruz.png')
files.download('cruz.png')

wordcloud = WordCloud().generate(shapiro)
# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.savefig('shapiro.png')
files.download('shapiro.png')
```
