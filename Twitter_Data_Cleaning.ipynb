```
# Import libraries
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
import nltk 
import string
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('words')


```

```
# Import data file
merged = pd.read_csv('/content/merged.csv')
merged.head()
# Subset by tweet and user for simple data analysis
merged = pd.DataFrame(merged)
tweets = merged[["username","tweet"]]
tweets = pd.DataFrame(tweets)
tweets.head()
```

```
# Remove @sign, http tags, splot, replace hashtags, http, image
# See https://stackoverflow.com/questions/64719706/cleaning-twitter-data-pandas-python for function reference, added image
def cleaner(tweet):
    tweet = re.sub("@[A-Za-z0-9]+","",tweet) #Remove @ sign
    tweet = re.sub(r"(?:\@|http?\://|https?\://|www)\S+", "", tweet) #Remove http links
    tweet = re.sub(r"(pic.twitter.com\/[^\s]+)", "", tweet) #Remove image links

    tweet = " ".join(tweet.split())
    tweet = tweet.replace("#", "").replace("_", " ") #Remove hashtag sign but keep the text
    return tweet
tweets['tweet'] = tweets['tweet'].map(lambda x: cleaner(x))
tweets.head()
```

```
#Next, prepare the data for analysis by normalizing, tokenizing, removing punctuation (Referenced https://medium.com/analytics-vidhya/a-simple-yet-effective-way-of-text-cleaning-using-nltk-4f90a8ff21d4)
cleantweet = []
text_data = ""
for i in range(0, len(tweets)):
  text_data = re.sub('[^a-zA-Z]', ' ', tweets['tweet'][i])
  text_data = text_data.lower()
  text_data = text_data.split()
  wl = WordNetLemmatizer()
  text_data = [wl.lemmatize(word) for word in text_data if not word in set(stopwords.words('english'))]
  text_data = ' '.join(text_data)
  cleantweet.append(text_data)
print(cleantweet)
tweets['tweet'] = cleantweet # Replace tweet dataframe
```

```
tweets.to_csv('clean.csv')
```
